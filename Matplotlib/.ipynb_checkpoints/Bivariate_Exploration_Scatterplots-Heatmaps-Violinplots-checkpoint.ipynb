```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
fuel_econ = pd.read_csv('fuel-econ.csv')
print(fuel_econ.shape)
fuel_econ.head(10)
```

# Quantitative vs Quantitative

```python
plt.scatter(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb')
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
sb.regplot(data = fuel_econ, x = 'displ', y = 'comb',
          x_jitter = 0.04, scatter_kws = {'alpha' : 1/10})
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb')
plt.colorbar()
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

```python
fuel_econ[['displ', 'comb']].describe()
```

```python
#By default vitidis colors are used, here I reverse that (as well as set a minimum of 0.5 so 0s are white)
# Based on the description table above, we should go with 0.3 bins for displ and 3 for comb
bins_x = np.arange(0.6, 7+0.3, 0.3)
bins_y = np.arange(12, 58+3, 3)
'''plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5,
           cmap = 'viridis_r', bins = [bins_x, bins_y]) '''
plt.hist2d(data = fuel_econ, x = 'displ', y = 'comb', cmin = 0.5,
           cmap = 'vlag_r', bins = [bins_x, bins_y]) 
plt.colorbar()
plt.xlabel('Displacement (1)')
plt.ylabel('Combined Fuel Efficiency (mpg)')
```

# Discrete vs Quantitative

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb')
```

```python
sb.regplot(data = fuel_econ, x = 'year', y = 'comb', x_jitter = 0.3, scatter_kws = {'alpha' : 1/20})
```

# Violin Plots

```python
sedan_classes = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
vclasses = pd.api.types.CategoricalDtype(ordered = True, categories = sedan_classes)
fuel_econ['VClass'] = fuel_econ['VClass'].astype(vclasses)

fuel_econ.head(10)
```

```python
base_color = sb.color_palette()[0]
sb.violinplot(data = fuel_econ, x = 'VClass', y = 'comb',
              color = base_color);
plt.xticks(rotation = 15);
```
